```python
# ---------------------------------------------------------
# EDA e AnÃ¡lise EstatÃ­stica
# ---------------------------------------------------------
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/air_quality_clean.csv")
print(f"âœ… Dataset carregado: {df.shape}")

# EstatÃ­sticas descritivas
print("\nðŸ“Š EstatÃ­sticas:")
print(df.describe().T)

# Top 10 cidades
plt.figure(figsize=(8,5))
sns.barplot(y=df['City'].value_counts().head(10).index,
            x=df['City'].value_counts().head(10).values,
            palette='crest')
plt.title("Top 10 cidades com mais registros")
plt.xlabel("Quantidade de registros")
plt.show()

# DistribuiÃ§Ã£o de faixas de temperatura
sns.countplot(y='Temp_Bin', data=df, palette='coolwarm',
              order=df['Temp_Bin'].value_counts().index)
plt.title("DistribuiÃ§Ã£o de Temperaturas")
plt.show()

# CorrelaÃ§Ã£o
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='RdBu_r', center=0)
plt.title("Matriz de CorrelaÃ§Ã£o")
plt.show()
```
